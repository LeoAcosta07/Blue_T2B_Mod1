## Mini Projeto 01 - Conversor de Moeda
Vamos construir um programa que irá converter moedas do real para o dólar e do dólar para o real. Vamos considerar que `$ 1,00 = R$ 5,75`

### Parte 1
Faça o conversor de moeda receber o valor em real e mostrar o valor convertido para dólar no formato `$ XXXX.XX`

**Exemplo:**

```
Valor em R$ = 1000
Valor em $ = $ 173.91
```

### Parte 2
Altere o conversor de moedas para receber o valor em dólar, converter para real e mostrar o resultado no formato `R$ XXXX.XX`

**Exemplo:**
```
Valor em $ = 1000
Valor em R$ = R$ 5750.00
```

````python
print("Bemvindo ao conversor")
print("1- Converter Real para Dolar")
print("2- Converter Dolar para Real")
moeda = input("Diguite a opcao que quer 1 ou 2:")
dolar = 5.75
#real1 = 1/dolar

if moeda == '1':
#conversao Real para Dolar
  conv_real = float(input("Digite a quantidade a converter: "))
  conv = conv_real/dolar
  print(f'O valor de R$ {conv_real:.2f} en Dolar é = U$S {conv:.2f}')

else:
#conversao Dolar para Real
  conv_dolar = float(input("Digite a quantidade a converter: "))
  conv2 = dolar*conv_dolar
  print(f'O valor de U$S {conv_dolar:.2f} en Real é = R$ {conv2:.2f}')
````
